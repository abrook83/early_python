Generate a NumPy array from the random seed set 101, with the dimesnions (4,3).

```python
myindex = ['CA','NY','AZ','TX']
```

```python
mycolumns = ['Jan','Feb','Mar']
```

```python
df = pd.DataFrame(data=mydata)
df
```

Now pass in the index values -

```python
df = pd.DataFrame(data=mydata,index=myindex)
df
```

And finally the columns -

```python
df = pd.DataFrame(data=mydata,index=myindex,columns=mycolumns)
df 
```

### Understanding File Paths

You have two options when reading a file with pandas:

1. If your .py file or .ipynb notebook is located in the **exact** same folder location as the .csv file you want to read, simply pass in the file name as a string, for example:
    
        df = pd.read_csv('some_file.csv')
        
2. Pass in the entire file path if you are located in a different directory. The file path must be 100% correct in order for this to work. For example:

        df = pd.read_csv("C:\\Users\\myself\\files\\some_file.csv")

# Index Basics

Before going over the same retrieval tasks for rows, let's build some basic understanding of the pandas DataFrame Index.

```python
df.head()
```

It is important that data is indexed with a unique value for each column. As we have not defined an index, one is automatically created with sequential numbers, thus ensuring each row has a unique value -

```python
df.index
```

Using 'set_index' we can use a column with unique individual values to set as the array's index ('payment_id' in this case) -

The index will now move to the left column, and is technically now not a column, but the index.

```python
df.set_index('Payment ID')
```

But the changes are not permanent....

```python
df.head()
```

So we can reassign it to the df variable name to make the changes permanent -

```python
df = df.set_index('Payment ID')
```

```python
df.head()
```

We can also reset the index to the range number values (0 - 244) -

```python
df = df.reset_index()
```

```python
df.head()
```
