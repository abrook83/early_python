## Grid Search

Grid searching allows us to apply multiple variations of hyperparameters to models to determine the best parameter combinations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/Advertising.csv")
```

```python
df.head()
```

Create the usual X & y, and scale the data -

```python
## CREATE X and y
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Model

```python
from sklearn.linear_model import ElasticNet
```

```python
help(ElasticNet)
```

With ElasticNet we provide two values;
- the alpha value, that multiplies the penalty terms dependant on its value between 0 & 1, effectively choosing somewhere between Ridge regression and Lasso, &
- the L1 ratio value which selects the amount with which the L1 & L2 parameters are mixed...

```python
base_elastic_net_model = ElasticNet()
```

Now we will create a python dictionary of values to be passed in and adjusted in our model. Using shift+tab above, we can see the available & adustable parameters in the ElasticNet model.

```python
param_grid = {'alpha':[0.1,5,10,50,100],
              'l1_ratio':[0.1,0.5,0.7,0.95,0.99,1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

When creating the grid search model, we must pass in:
- an estimator (regressor or classifier such as sklearn.svm.SVC());
- a parameter space;
- a method for searching or sampling candidates;
- a cross-validation scheme
- a score function.

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=2)
```

Now we'll take our model and fit the training sets to it -

Remember with this model we're effectively just performing cross validation where we're passing in a number of adjustable parameters to test them at the same step.

```python
grid_model.fit(X_train,y_train)
```

By running the 'best_estimator_' function we can extract which combination by our terms performed the best -

```python
grid_model.best_estimator_
```

...or just 'best_params_' to get just the values in a dictionary form -

```python
grid_model.best_params_
```

By running the 'cv_results_' into a DataFrame we can see a dictionary of how ALL the combinations performed to get an idea of how parameters have affected each other -

```python
pd.DataFrame(grid_model.cv_results_)
```

If we're satisfied with the above results, we can pass the test parameters into the model that will use its best results -

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error
```

```python
mse(y_test,y_pred)
```
